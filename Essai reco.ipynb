```python
# Create the TfidfVectorizer object with adjusted parameters
tfv = TfidfVectorizer(min_df=1, max_features=None, strip_accents='unicode', analyzer='word', token_pattern=r'\w{1,}', ngram_range=(1, 3), stop_words=['.',' ','and', 'the', 'a', 'an', 'in', 'of', 'to', 'is', 'it', 'that'])

# Transform the text data using the TfidfVectorizer
tfv_matrix = tfv.fit_transform(df[['genre_1', 'genre_2', 'genre_3', 'actor_1', 'actor_2', 'actor_3', 'plot', 'country', 'rated']])

# Check the shape of the tf-idf matrix
print(tfv_matrix.shape)
```

```python
indices = pd.Series(df.index, index = df.title).drop_duplicates()
indices
```

```python
def give_rec(title, sig, df):
    # Check if the title exists in the indices dictionary
    if title not in indices:
        return pd.Series()  # Return an empty Series if the title does not exist

    # Get the index corresponding to the title
    idx = indices[title]

    # Check if the index is within the bounds of the sig array
    if idx >= len(sig):
        return pd.Series()  # Return an empty Series if the index is out of bounds

    # Sort the movies
    # Selecting top 10 movies for recommendation
    sig_scores = list(enumerate(sig[idx]))
    sig_scores = sorted(sig_scores, key=lambda x: x[1], reverse=True)
    sig_scores = sig_scores[1:11]  # Exclude the first movie (itself)

    # Get the indices of the recommended movies
    movie_indices = [i[0] for i in sig_scores]

    # Get the titles of the recommended movies
    rec_movies = df['title'].iloc[movie_indices]

    return rec_movies
```

```python
# Importation des bibliothèques nécessaires
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
import pandas as pd


# Sélection des caractéristiques à utiliser pour la recommandation
features = df[['startYear',
               'runtimeMinutes', 'averageRating',
               'actor_1_factorized', 'actor_2_factorized', 'actor_3_factorized',
               'rated_factorized', 'primaryName_factorized', 'genre1_mapped',
               'genre2_mapped', 'genre3_mapped']]

y = df['title']

# Séparation des données en ensembles d'apprentissage et de test
X_train, X_test, y_train, y_test = train_test_split(features, y, test_size=0.2, random_state=42)

# Création de l'arbre de décision
decision_tree = DecisionTreeClassifier()

# Apprentissage de l'arbre de décision sur les données d'apprentissage
decision_tree.fit(X_train, y_train)

# Prédiction des recommandations pour tous les films
predictions = decision_tree.predict(features)

# Ajout des prédictions au dataframe
df['recommended'] = predictions
```

```python
from sklearn.metrics.pairwise import cosine_similarity
from sklearn.feature_extraction.text import TfidfVectorizer

# Créer une matrice TF-IDF pour les titres de films
tfidf = TfidfVectorizer()
tfidf_matrix = tfidf.fit_transform(df['title'].fillna(''))

# Calculer la similarité cosinus entre les films
cosine_sim = cosine_similarity(tfidf_matrix, tfidf_matrix)

# Fonction pour obtenir les recommandations en entrant un titre de film
def get_recommendations(title, cosine_sim=cosine_sim, df=df):
    # Obtenir l'index du film donné
    idx = df[df['title'] == title].index[0]

    # Obtenir les scores de similarité cosinus pour tous les films
    sim_scores = list(enumerate(cosine_sim[idx]))

    # Trier les films en fonction des scores de similarité
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)

    # Récupérer les 10 films les plus similaires
    sim_scores = sim_scores[1:11]
    movie_indices = [i[0] for i in sim_scores]

    # Retourner les titres des films recommandés
    return df['title'].iloc[movie_indices]


get_recommandations("Kate et Léopold")

```

```python
from sklearn.metrics.pairwise import pairwise_kernels

# Sélectionner les caractéristiques des films à utiliser
features = df[['startYear',
               'runtimeMinutes', 'averageRating',
               'actor_1_factorized', 'actor_2_factorized', 'actor_3_factorized',
               'rated_factorized', 'primaryName_factorized', 'genre1_mapped',
               'genre2_mapped', 'genre3_mapped']]

# Calculer la matrice de similarité à l'aide de la fonction sigmoïde de kernel
similarity_matrix = pairwise_kernels(features, metric='sigmoid')

def recommend_similar_movies(title, n=5):
    # Obtenir l'indice du film dans la matrice de similarité
    movie_index = df[df['title'] == title].index[0]
    # Obtenir les films les plus similaires
    similar_movies = similarity_matrix[movie_index].argsort()[::-1][:n+1]
    # Retourner les n films les plus similaires (à l'exception du film lui-même)
    return df.iloc[similar_movies[1:]]

recommend_similar_movies('Toy Story', n=5)
```
